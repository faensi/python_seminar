## ‚úèÔ∏è Teil 1: Grenzwerte herausfinden

**Was sind Grenzwerte?**

- Ein Grenzwert beschreibt, was mit einer Funktion passiert, wenn `x` gegen einen bestimmten Wert oder gegen unendlich l√§uft.
- Beispiel: 
  - \( \lim_{x \to \infty} \frac{1}{x} = 0 \)
  - \( \lim_{x \to 0^+} \frac{1}{x} = \infty \)
  

```python
# Beispiel: Grenzwert grafisch sichtbar machen
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 2, 100)
y = 1 / x

plt.plot(x, y)
plt.title('Grafische Ann√§herung an den Grenzwert von 1/x f√ºr x -> 0')
plt.xlabel('x')
plt.ylabel('1/x')
plt.grid(True)
plt.show()
```

### Berechnung von Grenzwerten mit `sympy`
Nutze `sympy`, um Grenzwerte symbolisch zu berechnen.

```python
%conda install sympy
```

```python
from sympy import Symbol, limit

x = Symbol('x')
expr = 1 / x

lim_0_plus = limit(expr, x, 0, '+')
lim_inf = limit(expr, x, float('inf'))

print(f'Grenzwert f√ºr x -> 0+: {lim_0_plus}')
print(f'Grenzwert f√ºr x -> ‚àû: {lim_inf}')
```

## ‚úèÔ∏è Teil 2: Regressionsanalyse

**Was ist Regressionsanalyse?**

- Ein Verfahren, um einen funktionalen Zusammenhang zwischen Variablen zu finden.
- Einfachste Form: **lineare Regression** \( y = m \cdot x + b \ )
- Hilft Trends zu erkennen und Vorhersagen zu treffen.

```python
# Einfaches Beispiel: Lineare Regression
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Regressionsanalyse
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f'Steigung: {slope:.2f}')
print(f'Achsenabschnitt: {intercept:.2f}')
print(f'Korrelationskoeffizient: {r_value:.2f}')

# Trendlinie
line = slope * x + intercept

plt.scatter(x, y, label='Datenpunkte')
plt.plot(x, line, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.title('Lineare Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
