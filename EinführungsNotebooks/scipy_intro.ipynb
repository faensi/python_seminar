# Was ist SciPy?

SciPy (Scientific Python) ist eine umfangreiche Sammlung von Algorithmen und mathematischen Werkzeugen, die auf NumPy aufbaut. Sie erweitert die Möglichkeiten von Python im Bereich des wissenschaftlichen Rechnens erheblich. <br>

Wichtige Merkmale von SciPy sind: <br>
- Umfassende Sammlung wissenschaftlicher Algorithmen <br>
- Optimierte und getestete Implementierungen mathematischer Routinen <br>
- Modulare Struktur für verschiedene Anwendungsbereiche <br>
- Nahtlose Integration mit NumPy <br>
- Aktive wissenschaftliche Community <br>

Die wichtigsten Module von SciPy sind: <br>
- **scipy.optimize**: Algorithmen für Optimierung und Wurzelsuche <br>
- **scipy.integrate**: Routinen für numerische Integration <br>
- **scipy.linalg**: Erweiterte lineare Algebra-Funktionen <br>
- **scipy.signal**: Signalverarbeitung und Filterdesign <br>
- **scipy.sparse**: Werkzeuge für dünn besetzte Matrizen <br>
- **scipy.stats**: Statistische Verteilungen und Funktionen <br>
- **scipy.interpolate**: Interpolationswerkzeuge <br>

SciPy ist besonders wichtig in: <br>
- Wissenschaftlicher Forschung <br>
- Ingenieurwesen <br>
- Datenanalyse <br>
- Signalverarbeitung <br>
- Bildverarbeitung <br>

Die Bibliothek ist zum Standard in der wissenschaftlichen Python-Community geworden, da sie zuverlässige, effiziente und gut dokumentierte Implementierungen wichtiger Algorithmen bietet. <br>

## NumPy vs. SciPy: Was ist der Unterschied?

NumPy und SciPy sind eng miteinander verbunden, haben aber unterschiedliche Schwerpunkte: <br>

### NumPy bietet:
- Grundlegende Array-Operationen <br>
- Elementweise mathematische Operationen <br>
- Einfache lineare Algebra <br>
- Basis-Funktionen für zufällige Zahlen <br>
- Grundlegende statistische Operationen <br>

### SciPy baut auf NumPy auf und bietet:
- Fortgeschrittene wissenschaftliche Algorithmen <br>
- Komplexe mathematische Routinen <br>
- Spezialisierte Anwendungen <br>

### Wann welche Bibliothek?

**NumPy, wenn:**
- Du hauptsächlich mit Arrays arbeitest <br>
- Du grundlegende mathematische Operationen brauchst <br>
- Du elementweise Berechnungen durchführst <br>
- Du einfache lineare Algebra benötigst <br>

**SciPy, wenn:**
- Du spezielle wissenschaftliche Algorithmen brauchst <br>
- Du numerische Integration durchführst <br>
- Du Optimierungsprobleme löst <br>
- Du fortgeschrittene statistische Tests benötigst <br>
- Du Signalverarbeitung implementierst <br>
- Du sparse Matrizen verarbeitest <br>

### Zusammenspiel:

In der Praxis werden beide Bibliotheken oft zusammen verwendet: <br>
- NumPy für die grundlegende Datenstruktur (Arrays) <br>
- SciPy für fortgeschrittene Berechnungen auf diesen Arrays <br>



## 1 Optimierung

SciPy bietet verschiedene Funktionen zur Optimierung: <br>

````python
from scipy import optimize
import numpy as np

# Funktion definieren die minimiert werden soll
def f(x):
    return (x[0] - 2)**2 + (x[1] - 3)**2

# Startpunkt
x0 = [0, 0]

# Minimierung
ergebnis = optimize.minimize(f, x0)
print("Minimierungsergebnis:")
print(f"x = {ergebnis.x[0]:.2f}, y = {ergebnis.x[1]:.2f}")
print(f"Minimaler Wert = {ergebnis.fun:.2f}")

# Wurzelsuche
def g(x):
    return x**2 - 4

wurzel = optimize.root(g, x0=3)
print("\nWurzel von x²-4:", wurzel.x[0])
````

## 2 Integration und Differentiation

Numerische Integration und Differentiation: <br>

````python
from scipy import integrate

# Integration
def h(x):
    return np.sin(x)**2

# Bestimmtes Integral
ergebnis, fehler = integrate.quad(h, 0, np.pi)
print(f"Integral von sin²(x) von 0 bis π = {ergebnis:.6f}")

# Mehrfachintegration
def f2(x, y):
    return x * y

ergebnis, fehler = integrate.dblquad(f2, 0, 1,    # y-Grenzen
                                    lambda x: 0,    # x-Untergrenze
                                    lambda x: 1)    # x-Obergrenze
print(f"\nDoppelintegral = {ergebnis:.6f}")
````

## 3 Lineare Algebra

Erweiterte lineare Algebra-Funktionen: <br>

````python
from scipy import linalg

# Matrix erstellen
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 6])

# Lineares Gleichungssystem lösen
x = linalg.solve(A, b)
print("Lösung des Gleichungssystems:")
print(x)

# Eigenwerte und Eigenvektoren
eigvals, eigvecs = linalg.eig(A)
print("\nEigenwerte:")
print(eigvals)

# LU-Zerlegung
P, L, U = linalg.lu(A)
print("\nLU-Zerlegung - U-Matrix:")
print(U)
````

## 4 Signalverarbeitung

Das Signal-Modul für Signalverarbeitung: <br>

````python
from scipy import signal
import numpy as np

# Signal generieren
t = np.linspace(0, 1, 1000)
x = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*10*t)

# Filter entwerfen
b, a = signal.butter(4, 0.2, 'low')
y = signal.filtfilt(b, a, x)

# Spektralanalyse
f, Pxx = signal.welch(x, fs=1000)

# Fensterfunktionen
fenster = signal.windows.hann(50)
print("Hann-Fenster (erste 5 Werte):")
print(fenster[:5])
````

## 5 Statistik

Statistische Funktionen und Tests: <br>

````python
from scipy import stats

# Normalverteilte Daten generieren
daten = np.random.normal(loc=0, scale=1, size=1000)

# Deskriptive Statistik
print("Normalverteilungstest:")
print(stats.normaltest(daten))

# t-Test
gruppe1 = np.random.normal(0, 1, 100)
gruppe2 = np.random.normal(0.5, 1, 100)
t_stat, p_wert = stats.ttest_ind(gruppe1, gruppe2)
print("\nt-Test p-Wert:", p_wert)

# Korrelation
x = np.random.normal(0, 1, 100)
y = 2*x + np.random.normal(0, 0.5, 100)
r, p = stats.pearsonr(x, y)
print("\nPearson Korrelation:", r)
````

## 6 Interpolation

Verschiedene Interpolationsmethoden: <br>

````python
from scipy import interpolate

# Datenpunkte
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 2, 1, 3, 7, 4])

# Verschiedene Interpolationen
f_linear = interpolate.interp1d(x, y)
f_cubic = interpolate.interp1d(x, y, kind='cubic')
f_spline = interpolate.UnivariateSpline(x, y)

# Neue Punkte für glatte Kurve
x_new = np.linspace(0, 5, 100)
y_linear = f_linear(x_new)
y_cubic = f_cubic(x_new)
y_spline = f_spline(x_new)

print("Interpolierte Werte bei x=2.5:")
print(f"Linear: {f_linear(2.5):.2f}")
print(f"Kubisch: {f_cubic(2.5):.2f}")
print(f"Spline: {f_spline(2.5):.2f}")
````

## 7 Sparse Matrizen

Effiziente Handhabung dünn besetzter Matrizen: <br>

````python
from scipy import sparse

# Sparse Matrix erstellen
data = np.array([1, 2, 3, 4])
row = np.array([0, 1, 2, 2])
col = np.array([0, 1, 2, 1])
sparse_matrix = sparse.csr_matrix((data, (row, col)), shape=(3, 3))

print("Sparse Matrix als Array:")
print(sparse_matrix.toarray())

# Operationen
dicht = sparse_matrix @ sparse_matrix.T
print("\nMatrixprodukt:")
print(dicht.toarray())
````

## 2. scipy.optimize - Optimierungsmethoden

Das `scipy.optimize`-Modul bietet Funktionen zur Minimierung (oder Maximierung) von Funktionen, für univariate und multivariate Probleme, mit oder ohne Beschränkungen.

````python
import numpy as np
import matplotlib.pyplot as plt
````

### Hauptfunktionen

- `minimize`: Allgemeine Minimierungsfunktion
- `curve_fit`: Anpassung einer Kurve an Datenpunkte
- `root`: Nullstellensuche
- `linprog`: Lineare Programmierung
- `minimize_scalar`: Minimierung einer eindimensionalen Funktion

### Beispiel 1: Funktionsminimierung

````python
from scipy import optimize


# Funktion, die minimiert werden soll
def f(x):
    return x**2 + 10*np.sin(x)

# Visualisierung der Funktion
x = np.linspace(-10, 10, 1000)
plt.plot(x, f(x))
plt.grid(True)
plt.title('f(x) = x² + 10sin(x)')
plt.show()

# Minimierung der Funktion
result = optimize.minimize(f, x0=0)  # Startwert x0=0
print("Minimum gefunden bei x =", result.x[0])
print("Funktionswert am Minimum:", result.fun)
print("Erfolgreich:", result.success)
print("Anzahl der Funktionsaufrufe:", result.nfev)
````

### Beispiel 2: Kurvenanpassung (Curve Fitting)

````python

import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

# Echte Funktion, die wir anpassen wollen
def echte_funktion(x, a, b, c):
    return a * np.exp(-b * x) + c

# Generiere synthetische Daten mit Rauschen
x_daten = np.linspace(0, 4, 50)
y_daten = echte_funktion(x_daten, 2.5, 1.3, 0.5) + 0.2 * np.random.normal(size=len(x_daten))

# Kurvenanpassung durchführen
params, params_covariance = optimize.curve_fit(echte_funktion, x_daten, y_daten)

print("Geschätzte Parameter:")
print(f"a = {params[0]:.4f}")
print(f"b = {params[1]:.4f}")
print(f"c = {params[2]:.4f}")

# Visualisierung
plt.scatter(x_daten, y_daten, label='Daten')
plt.plot(x_daten, echte_funktion(x_daten, *params), 'r-', label='Angepasste Kurve')
plt.legend()
plt.show()
````

### Beispiel 3: Nullstellensuche

````python
```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

# Funktion, deren Nullstellen wir finden wollen
def g(x):
    return x**3 - 2*x**2 - 11*x + 12

# Visualisierung
x = np.linspace(-5, 5, 1000)
plt.plot(x, g(x))
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.title('g(x) = x³ - 2x² - 11x + 12')
plt.show()

# Nullstellensuche mit verschiedenen Startwerten
nullstelle1 = optimize.root(g, x0=-4)
nullstelle2 = optimize.root(g, x0=0)
nullstelle3 = optimize.root(g, x0=4)

print("Nullstellen:")
print(f"x₁ = {nullstelle1.x[0]:.6f}")
print(f"x₂ = {nullstelle2.x[0]:.6f}")
print(f"x₃ = {nullstelle3.x[0]:.6f}")

# Überprüfung
for i, ns in enumerate([nullstelle1.x[0], nullstelle2.x[0], nullstelle3.x[0]]):
    print(f"g(x₁) = {g(ns):.10f}")
```

### Übungsaufgaben zu scipy.optimize

#### Übung 1: Einfache Minimierung
Finden Sie das Minimum der Funktion `f(x) = x^4 - 4x^2 + 5` im Intervall [-10, 10].
   - Visualisieren Sie die Funktion
   - Verwenden Sie `minimize_scalar` mit verschiedenen Methoden ('brent', 'golden', 'bounded')
   - Vergleichen Sie die Ergebnisse

#### Übung 2: Multivariate Optimierung
Finden Sie das Minimum der Rosenbrock-Funktion: `f(x,y) = (1-x)^2 + 100*(y-x^2)^2`
   - Verwenden Sie `minimize` mit verschiedenen Algorithmen (z.B. 'BFGS', 'Nelder-Mead')
   - Starten Sie von verschiedenen Punkten, z.B. (-1, 1), (0, 0), (2, 2)
   - Visualisieren Sie die Konvergenz für verschiedene Algorithmen

#### Übung 3: Kurvenanpassung mit eigenen Daten
Erstellen Sie einen Datensatz, der einem Polynom dritten Grades mit Rauschen folgt.
   - Generieren Sie 50 Datenpunkte im Bereich [-5, 5]
   - Fügen Sie normalverteiltes Rauschen hinzu
   - Verwenden Sie `curve_fit`, um die Koeffizienten des Polynoms zu schätzen
   - Visualisieren Sie die Daten und die angepasste Kurve

#### Übung 4: Constrained Optimization
Lösen Sie folgendes Optimierungsproblem mit Nebenbedingungen:
   - Minimiere: `f(x,y) = (x-1)^2 + (y-2.5)^2`
   - Unter den Nebenbedingungen: `x + y ≤ 1`, `x ≥ 0`, `y ≥ 0`
   - Verwenden Sie `minimize` mit der Methode 'SLSQP' und definieren Sie die Nebenbedingungen
   - Visualisieren Sie das Ergebnis

#### Übung 5: Root Finding mit System von Gleichungen
Lösen Sie das folgende nichtlineare Gleichungssystem:
   - `f₁(x,y) = x^2 + y^2 - 1 = 0`
   - `f₂(x,y) = x^2 - y - 0.5 = 0`
   - Verwenden Sie `root` mit einer geeigneten Startschätzung
   - Visualisieren Sie beide Gleichungen und die gefundene Lösung
````
